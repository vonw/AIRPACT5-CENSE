```python
wrf = rand(285,258) * 30.
```

```python
pcolormesh(wrf)
colorbar()
```

```python
conditions = np.arange(0,7)
ages       = np.arange(1,112)
```

```python
for condition in conditions:
    relativeRisk = np.empty(shape=(1,285,258))
    for age in ages:
        # print(condition, age)
        relativeRisk = np.concatenate((relativeRisk, np.reshape(rand(1) * 0.01 * wrf, (1,285,258))))
    relativeRisk = relativeRisk[1:,:,:]
    print('Writing file for: Condition' + str(condition))
    np.save('Condition' + str(condition), relativeRisk)
```
